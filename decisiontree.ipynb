```python
import numpy as np
import pandas as pd
from sklearn import tree
```

```python
restaurantDataset = pd.read_csv('data_restaurant2.csv',delimiter=',',header=0)
```

```python
restaurantDataset.head()
```

```python
restaurantDataset["Restaurant"] = pd.factorize(restaurantDataset.Restaurant)[0]
```

```python
restaurantDataset.head()
```

```python
restaurantDataset = restaurantDataset.drop(labels="Rank",axis=1)
```

```python
print(restaurantDataset)
```

```python
restaurantDataset = restaurantDataset.to_numpy()
```

```python
print(restaurantDataset)
```

```python
dataTraining = np.concatenate((restaurantDataset[0:5, :], restaurantDataset[10:15, :]),axis=0)
dataTesting = np.concatenate((restaurantDataset[40:45, :], restaurantDataset[20:25, :]),axis=0)
```

```python
print(dataTesting)
len(dataTesting)
```

```python
print(dataTraining)
len(dataTraining)
```

```python
inputTraining = dataTraining[:,0:4]
inputTesting = dataTesting[:,0:4]
labelTraining = dataTraining[:,4]
labelTesting = dataTesting[:,4]
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model = model.fit(inputTraining, labelTraining)
```

```python
hasilPrediksi = model.predict(inputTesting)
```

```python
print("Label sebenarnya ",labelTesting)
print("Hasil prediksi ",hasilPrediksi)
```

```python
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
```

```python
print("Prediksi benar: ", prediksiBenar, " data")
print("Prediksi salah: ", prediksiSalah, " data")
print("Akurasi: ",prediksiBenar/(prediksiBenar+prediksiSalah)*100,"%")
```
